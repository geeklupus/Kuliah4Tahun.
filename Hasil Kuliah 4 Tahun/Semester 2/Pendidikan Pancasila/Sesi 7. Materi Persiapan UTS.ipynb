Jawaban

Berikut adalah penjelasan mengenai keterkaitan antara konsep dan urgensi pendidikan Pancasila dalam menghadapi disintegrasi bangsa serta sejarah tersusunnya dasar negara Pancasila, termasuk para tokoh yang berjasa dalam penyusunannya.

### 1. Keterkaitan antara Konsep dan Urgensi Pendidikan Pancasila dalam Menghadapi Disintegrasi Bangsa

**Pendidikan Pancasila** sangat penting dalam menghadapi tantangan disintegrasi bangsa yang terjadi saat ini. Disintegrasi bangsa mengacu pada potensi perpecahan yang disebabkan oleh berbagai faktor seperti konflik sosial, perbedaan ideologi, ketidakadilan sosial, separatisme, dan radikalisme. Berikut adalah keterkaitan antara konsep dan urgensi pendidikan Pancasila dalam menghadapi disintegrasi bangsa:

#### a. **Konsep Pendidikan Pancasila**

Pendidikan Pancasila adalah proses pendidikan yang bertujuan untuk menginternalisasikan nilai-nilai Pancasila dalam kehidupan sehari-hari masyarakat Indonesia. Pendidikan ini tidak hanya mencakup pemahaman teoritis tentang sila-sila Pancasila, tetapi juga penerapan praktis dari nilai-nilai tersebut dalam kehidupan sosial, politik, ekonomi, dan budaya.

#### b. **Urgensi Pendidikan Pancasila dalam Menghadapi Disintegrasi Bangsa**

1. **Membangun Kesadaran Kolektif tentang Keberagaman dan Persatuan**
   - **Menghargai Keberagaman**: Pendidikan Pancasila membantu masyarakat memahami dan menghargai keberagaman suku, agama, ras, dan golongan di Indonesia. Ini penting untuk mencegah diskriminasi dan konflik horizontal yang dapat mengancam persatuan bangsa.
   - **Menumbuhkan Semangat Persatuan**: Melalui pendidikan Pancasila, nilai-nilai persatuan dan kesatuan bangsa dapat ditanamkan sejak dini, sehingga masyarakat lebih menghargai dan menjaga keutuhan negara dalam menghadapi perbedaan.

2. **Mengatasi Radikalisme dan Ekstremisme**
   - **Menangkal Ideologi Radikal**: Pendidikan Pancasila dapat berfungsi sebagai benteng untuk menangkal masuknya ideologi radikal dan ekstrem yang bertentangan dengan nilai-nilai kebangsaan. Dengan memahami Pancasila, masyarakat diharapkan memiliki imunitas ideologis terhadap paham-paham yang dapat memecah belah bangsa.
   - **Meningkatkan Toleransi dan Harmoni Sosial**: Dengan memahami dan menerapkan nilai-nilai Pancasila, terutama yang terkait dengan kemanusiaan dan persatuan, masyarakat akan lebih toleran dan mampu hidup harmonis dalam keberagaman.

3. **Memperkuat Identitas dan Nasionalisme**
   - **Membentuk Identitas Nasional yang Kuat**: Pendidikan Pancasila membantu membentuk identitas nasional yang kuat di tengah arus globalisasi yang dapat mengikis rasa cinta tanah air. Ini penting untuk menjaga kedaulatan dan integritas negara dari ancaman disintegrasi.
   - **Menumbuhkan Rasa Bangga dan Cinta Tanah Air**: Dengan memahami sejarah dan filosofi di balik Pancasila, generasi muda dapat memiliki rasa bangga dan cinta yang mendalam terhadap tanah air, yang pada gilirannya akan mendorong mereka untuk menjaga keutuhan bangsa.

4. **Mendorong Partisipasi Aktif dalam Demokrasi**
   - **Mendidik Warga Negara yang Kritis dan Bertanggung Jawab**: Pendidikan Pancasila mendorong partisipasi aktif dalam kehidupan demokrasi, termasuk pemahaman tentang hak dan kewajiban sebagai warga negara. Partisipasi ini penting untuk memastikan bahwa setiap warga negara merasa memiliki dan bertanggung jawab terhadap keberlanjutan negara.
   - **Mengajarkan Prinsip Demokrasi yang Berkeadilan**: Melalui pendidikan Pancasila, masyarakat diajarkan untuk memahami demokrasi bukan hanya sebagai proses pemilu, tetapi juga sebagai praktik sosial yang berkeadilan, di mana setiap suara dihargai dan setiap keputusan diambil untuk kepentingan bersama.

#### c. **Solusi dalam Memperkuat Keutuhan Bangsa dan Negara**

1. **Mengintegrasikan Pendidikan Pancasila ke dalam Kurikulum Pendidikan**
   - Pendidikan Pancasila harus diintegrasikan ke dalam kurikulum pendidikan formal di semua jenjang, dari sekolah dasar hingga perguruan tinggi. Pendidikan ini harus dirancang sedemikian rupa agar relevan dan menarik bagi generasi muda.
   
2. **Meningkatkan Pendidikan Non-Formal dan Informal**
   - Selain pendidikan formal, perlu juga dikembangkan program pendidikan non-formal dan informal yang dapat menjangkau masyarakat luas, termasuk melalui media sosial, seminar, workshop, dan kegiatan komunitas lainnya.

3. **Menerapkan Program Edukasi yang Interaktif dan Kontekstual**
   - Pendidikan Pancasila harus disampaikan dengan cara yang interaktif dan kontekstual, menggunakan teknologi digital dan pendekatan yang relevan dengan tantangan zaman, sehingga lebih mudah dipahami dan diterapkan oleh masyarakat.

4. **Membangun Kesadaran Bersama tentang Ancaman Disintegrasi**
   - Melalui kampanye publik dan program edukatif, masyarakat harus disadarkan tentang ancaman disintegrasi bangsa dan pentingnya menjaga keutuhan negara melalui penerapan nilai-nilai Pancasila.

5. **Mendorong Dialog dan Komunikasi Antarbudaya**
   - Membangun dialog antarbudaya dan antaragama untuk memperkuat kesepahaman dan mengurangi potensi konflik yang dapat menyebabkan disintegrasi.

### 2. Sejarah Tersusunnya Dasar Negara Pancasila

Sejarah lahirnya Pancasila sebagai dasar negara Indonesia melibatkan berbagai peristiwa penting dan kontribusi dari banyak tokoh nasional. Berikut adalah rangkaian peristiwa dan peran tokoh dalam penyusunan Pancasila:

#### a. **Awal Mula Gagasan Pancasila**

1. **Sidang BPUPKI (Badan Penyelidik Usaha-usaha Persiapan Kemerdekaan Indonesia)**
   - Pada tanggal 29 Mei 1945, BPUPKI mengadakan sidang pertama untuk membahas dasar negara Indonesia yang merdeka. Sidang ini dihadiri oleh tokoh-tokoh nasional yang berasal dari berbagai latar belakang, baik etnis, agama, maupun ideologi.

2. **Pidato Soekarno tentang Pancasila (1 Juni 1945)**
   - Pada tanggal 1 Juni 1945, Soekarno menyampaikan pidato yang kemudian dikenal sebagai "Lahirnya Pancasila". Dalam pidato ini, Soekarno mengusulkan lima dasar negara yang meliputi:
     1. **Kebangsaan Indonesia** (Nasionalisme)
     2. **Internasionalisme atau Perikemanusiaan** (Kemanusiaan yang Adil dan Beradab)
     3. **Mufakat atau Demokrasi**
     4. **Kesejahteraan Sosial**
     5. **Ketuhanan Yang Maha Esa**

#### b. **Pembentukan Panitia Sembilan dan Piagam Jakarta**

1. **Pembentukan Panitia Sembilan**
   - Untuk menyempurnakan konsep dasar negara, BPUPKI membentuk Panitia Sembilan yang terdiri dari sembilan tokoh penting: Soekarno, Mohammad Hatta, Achmad Subardjo, Muhammad Yamin, Abdul Kahar Muzakkir, Abikusno Tjokrosuyoso, Agus Salim, Wahid Hasyim, dan A.A. Maramis.
   - Tugas Panitia Sembilan adalah merumuskan konsep dasar negara yang dapat diterima oleh semua golongan dan kelompok.

2. **Penyusunan Piagam Jakarta (22 Juni 1945)**
   - Pada tanggal 22 Juni 1945, Panitia Sembilan berhasil merumuskan Piagam Jakarta yang berisi rancangan pembukaan Undang-Undang Dasar (UUD) dengan lima prinsip dasar negara yang kemudian dikenal sebagai Pancasila:
     1. Ketuhanan dengan kewajiban menjalankan syariat Islam bagi pemeluk-pemeluknya.
     2. Kemanusiaan yang adil dan beradab.
     3. Persatuan Indonesia.
     4. Kerakyatan yang dipimpin oleh hikmat kebijaksanaan dalam permusyawaratan/perwakilan.
     5. Keadilan sosial bagi seluruh rakyat Indonesia.

#### c. **Perubahan dan Penyempurnaan Pancasila**

1. **Sidang PPKI (Panitia Persiapan Kemerdekaan Indonesia)**
   - Setelah proklamasi kemerdekaan pada 17 Agustus 1945, PPKI mengadakan sidang pada 18 Agustus 1945 untuk mengesahkan UUD 1945. Dalam sidang ini, terdapat diskusi mengenai perubahan sila pertama dari Piagam Jakarta.
   - Atas pertimbangan untuk menjaga persatuan dan tidak menimbulkan perpecahan, frasa "dengan kewajiban menjalankan syariat Islam bagi pemeluk-pemeluknya" diubah menjadi "Ketuhanan Yang Maha Esa" untuk menghormati kebhinekaan agama di Indonesia.

2. **Pengesahan Pancasila sebagai Dasar Negara (18 Agustus 1945)**
   - Pada 18 Agustus 1945, PPKI mengesahkan Pancasila sebagai dasar negara Indonesia yang terdiri dari lima sila dalam Pembukaan UUD 1945:
     1. Ketuhanan Yang Maha Esa.
     2. Kemanusiaan yang adil dan beradab.
     3. Persatuan Indonesia.
     4. Kerakyatan yang dipimpin oleh hikmat kebijaksanaan

 dalam permusyawaratan/perwakilan.
     5. Keadilan sosial bagi seluruh rakyat Indonesia.

#### d. **Para Tokoh yang Berjasa dalam Penyusunan Pancasila**

1. **Soekarno**: Mengusulkan konsep Pancasila melalui pidato pada 1 Juni 1945 yang menjadi landasan bagi penyusunan dasar negara.
2. **Mohammad Hatta**: Wakil ketua PPKI yang berperan dalam menyempurnakan Pancasila dan mengesahkan UUD 1945.
3. **Muhammad Yamin**: Salah satu anggota BPUPKI yang aktif dalam diskusi tentang dasar negara dan anggota Panitia Sembilan.
4. **Achmad Subardjo**: Anggota BPUPKI dan Panitia Sembilan yang turut berkontribusi dalam penyusunan Piagam Jakarta.
5. **Abikusno Tjokrosuyoso**: Anggota Panitia Sembilan yang mewakili kelompok Islam dan turut menyusun Piagam Jakarta.
6. **Agus Salim**: Anggota Panitia Sembilan yang mewakili kelompok Islam dan berperan dalam perumusan Piagam Jakarta.
7. **Wahid Hasyim**: Tokoh NU yang juga anggota Panitia Sembilan dan berperan dalam negosiasi untuk mengubah sila pertama Piagam Jakarta.
8. **A.A. Maramis**: Perwakilan Kristen dalam Panitia Sembilan yang turut menyepakati perubahan sila pertama demi menjaga persatuan bangsa.

### Kesimpulan

Pendidikan Pancasila memiliki peran penting dalam menghadapi disintegrasi bangsa dengan membangun kesadaran tentang keberagaman, menangkal radikalisme, memperkuat identitas nasional, dan mendorong partisipasi aktif dalam demokrasi. Sejarah penyusunan Pancasila menunjukkan kontribusi besar dari berbagai tokoh nasional yang berusaha menyusun dasar negara yang dapat diterima oleh semua golongan. Oleh karena itu, memahami dan menerapkan nilai-nilai Pancasila menjadi kunci untuk menjaga keutuhan bangsa dan negara Indonesia.
