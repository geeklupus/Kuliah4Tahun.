### 1. Sejarah Lahirnya Pancasila

Pancasila adalah dasar negara Republik Indonesia yang menjadi panduan dalam kehidupan berbangsa dan bernegara. Sejarah lahirnya Pancasila tidak lepas dari perjuangan bangsa Indonesia untuk merdeka dari penjajahan dan membangun sebuah negara yang berdaulat. Berikut adalah sejarah singkat lahirnya Pancasila:

1. **Awal Perumusan:**
   - Proses perumusan Pancasila dimulai ketika Jepang menjanjikan kemerdekaan kepada Indonesia pada masa pendudukan mereka selama Perang Dunia II. Pada 1 Maret 1945, Jepang mengumumkan pembentukan **Badan Penyelidik Usaha-usaha Persiapan Kemerdekaan Indonesia (BPUPKI)** untuk mempersiapkan segala sesuatu yang diperlukan bagi kemerdekaan Indonesia.
   
2. **Sidang Pertama BPUPKI (29 Mei - 1 Juni 1945):**
   - BPUPKI mengadakan sidang pertama dari 29 Mei hingga 1 Juni 1945 untuk membahas dasar negara Indonesia merdeka. Dalam sidang ini, beberapa tokoh memberikan pandangan mengenai dasar negara, di antaranya **Muhammad Yamin**, **Soepomo**, dan **Soekarno**.
   - **Soekarno** menyampaikan pidatonya pada 1 Juni 1945 dan mengusulkan konsep dasar negara yang disebutnya sebagai **Pancasila**. Dalam pidatonya, Soekarno mengusulkan lima prinsip dasar yaitu:
     1. Kebangsaan Indonesia
     2. Internasionalisme atau Perikemanusiaan
     3. Mufakat atau Demokrasi
     4. Kesejahteraan Sosial
     5. Ketuhanan yang Maha Esa

3. **Panitia Sembilan dan Piagam Jakarta:**
   - Setelah sidang pertama BPUPKI, dibentuk **Panitia Kecil** dan **Panitia Sembilan** yang bertugas menyusun rumusan Pancasila secara lebih konkret. **Panitia Sembilan** ini diketuai oleh Soekarno dan terdiri dari berbagai tokoh, seperti Muhammad Hatta, Ahmad Subardjo, Abikusno Tjokrosuyoso, Abdul Kahar Muzakkir, Agus Salim, Muhammad Yamin, Wahid Hasyim, dan AA Maramis.
   - Pada 22 Juni 1945, Panitia Sembilan menghasilkan sebuah rumusan yang dikenal sebagai **Piagam Jakarta** yang berisi lima sila sebagai dasar negara Indonesia merdeka.

4. **Sidang Kedua BPUPKI (10 - 17 Juli 1945):**
   - Dalam sidang kedua BPUPKI, Piagam Jakarta diterima sebagai dasar rancangan pembukaan Undang-Undang Dasar (UUD). Pada sidang ini juga dibahas pasal-pasal lain yang akan dimasukkan dalam UUD.

5. **Perubahan Piagam Jakarta dan Pengesahan Pancasila:**
   - Pada 17 Agustus 1945, Indonesia memproklamasikan kemerdekaannya. Namun, pada 18 Agustus 1945, sehari setelah proklamasi, **Panitia Persiapan Kemerdekaan Indonesia (PPKI)** mengadakan sidang untuk mengesahkan UUD 1945 dan Pancasila sebagai dasar negara.
   - Dalam sidang ini, terjadi perubahan pada sila pertama Pancasila dari "Ketuhanan dengan kewajiban menjalankan syariat Islam bagi pemeluk-pemeluknya" menjadi "Ketuhanan Yang Maha Esa" untuk mengakomodasi keberagaman agama di Indonesia.
   - Pada akhirnya, rumusan Pancasila seperti yang dikenal sekarang ini diresmikan sebagai dasar negara pada sidang PPKI tersebut.

### 2. Sumber Sosiologis dan Politis Pancasila dalam Kajian Sejarah Bangsa Indonesia

**Sumber Sosiologis Pancasila:**

Sumber sosiologis Pancasila berhubungan dengan nilai-nilai sosial, budaya, dan tradisi yang telah ada dan berkembang dalam masyarakat Indonesia sebelum Pancasila dirumuskan. Nilai-nilai ini mencerminkan karakteristik kehidupan masyarakat Indonesia yang penuh dengan keragaman namun tetap hidup dalam semangat persatuan. Berikut beberapa poin penting mengenai sumber sosiologis Pancasila:

1. **Kearifan Lokal dan Nilai-Nilai Budaya:**
   - Pancasila menggali nilai-nilai luhur yang telah hidup dalam masyarakat Indonesia, seperti gotong royong, musyawarah untuk mufakat, dan toleransi antarumat beragama. Nilai-nilai ini sudah menjadi bagian dari kehidupan masyarakat Indonesia sejak zaman kerajaan hingga masa kolonial.

2. **Pluralisme Masyarakat Indonesia:**
   - Indonesia adalah negara dengan keberagaman suku, agama, ras, dan golongan. Pancasila mengakomodasi pluralisme ini dengan menempatkan sila-sila yang mencerminkan semangat persatuan, toleransi, dan kemanusiaan, sehingga dapat menyatukan seluruh komponen bangsa dalam satu bingkai Negara Kesatuan Republik Indonesia (NKRI).

3. **Prinsip Gotong Royong:**
   - Prinsip gotong royong adalah ciri khas masyarakat Indonesia yang tercermin dalam berbagai aspek kehidupan sosial. Prinsip ini tercermin dalam sila keempat Pancasila yang menekankan pada musyawarah dan mufakat sebagai cara untuk mengambil keputusan bersama.

**Sumber Politis Pancasila:**

Sumber politis Pancasila berkaitan dengan situasi politik dan perjuangan bangsa Indonesia dalam merebut kemerdekaan serta membentuk negara yang merdeka, bersatu, berdaulat, adil, dan makmur. Beberapa poin penting mengenai sumber politis Pancasila:

1. **Perjuangan Kemerdekaan:**
   - Pancasila lahir dari semangat perjuangan bangsa Indonesia melawan penjajahan untuk mencapai kemerdekaan. Keinginan untuk membentuk suatu negara yang merdeka dan berdaulat tercermin dalam seluruh proses perumusan Pancasila.

2. **Perdebatan Politik di BPUPKI:**
   - Perumusan Pancasila juga merupakan hasil dari perdebatan politik yang panjang di BPUPKI dan PPKI, di mana para pendiri bangsa berusaha merumuskan dasar negara yang dapat diterima oleh semua golongan dan mencerminkan kepentingan nasional yang luas.

3. **Kompromi Politik:**
   - Pancasila adalah hasil dari kompromi politik antara kelompok nasionalis, kelompok agama, dan kelompok lainnya. Hal ini terlihat dari perubahan pada sila pertama Piagam Jakarta yang diubah menjadi "Ketuhanan Yang Maha Esa" untuk memastikan persatuan dan keutuhan bangsa Indonesia.

### 3. Isi Piagam Jakarta

**Piagam Jakarta**, yang dirumuskan pada 22 Juni 1945, adalah dokumen bersejarah yang menjadi landasan awal bagi dasar negara Indonesia sebelum disempurnakan menjadi Pancasila seperti yang kita kenal sekarang. Berikut adalah isi Piagam Jakarta:

1. Ketuhanan, dengan kewajiban menjalankan syariat Islam bagi pemeluk-pemeluknya.
2. Kemanusiaan yang adil dan beradab.
3. Persatuan Indonesia.
4. Kerakyatan yang dipimpin oleh hikmat kebijaksanaan dalam permusyawaratan/perwakilan.
5. Keadilan sosial bagi seluruh rakyat Indonesia.

Setelah perubahan pada sidang PPKI tanggal 18 Agustus 1945, sila pertama diubah menjadi "Ketuhanan Yang Maha Esa" untuk menghormati keragaman agama dan kepercayaan yang ada di Indonesia. Inilah rumusan yang akhirnya menjadi dasar negara Republik Indonesia.
