### 1. Diketahui a=(1,2), b=(-2,-3) dan e=(1,3). Apakah c merupakan kombinasi linier dari a dan b?

#### Langkah-langkah:
1. **Tentukan Vektor C**:
   - Vektor c belum diberikan secara eksplisit, namun kita dapat asumsikan bahwa kita ingin mengetahui apakah vektor c dapat diekspresikan sebagai kombinasi linier dari vektor a dan b.

2. **Kriteria Kombinasi Linier**:
   - Untuk menentukan apakah vektor c merupakan kombinasi linier dari vektor a dan b, kita harus mengekspresikan vektor c sebagai linear kombinasi dari a dan b.
   - Rumusnya adalah:
   $$
   c = \alpha a + \beta b
   $$
   - Di sini, α dan β adalah skalar.

3. **Contoh Vektor C**:
   - Misalkan kita ingin mengetahui apakah vektor c = (c1, c2) dapat diekspresikan sebagai kombinasi linier dari a dan b.
   - Maka kita harus menyelesaikan sistem persamaan:
   $$
   c_1 = \alpha (1) + \beta (-2)
   $$
   $$
   c_2 = \alpha (2) + \beta (-3)
   $$
   - Jika ada nilai α dan β yang memenuhi kedua persamaan di atas, maka c merupakan kombinasi linier dari a dan b.

#### Contoh:
- Misalkan kita ingin mengetahui apakah c = (0, 0) merupakan kombinasi linier dari a dan b.
- Maka kita harus menyelesaikan sistem persamaan:
  $$
  0 = \alpha (1) + \beta (-2)
  $$
  $$
  0 = \alpha (2) + \beta (-3)
  $$
  - Dari persamaan pertama, kita dapat menulis:
    $$
    0 = \alpha - 2\beta
    $$
    - Dari persamaan kedua, kita dapat menulis:
      $$
      0 = 2\alpha - 3\beta
      $$
    - Menggabungkan kedua persamaan, kita dapat menyelesaikan sistem persamaan:
      $$
      0 = 2\alpha - 3\beta
      $$
      $$
      0 = \alpha - 2\beta
      $$
    - Menggabungkan kedua persamaan, kita dapat menyelesaikan sistem persamaan:
      $$
      0 = 3\alpha - 5\beta
      $$
    - Jika tidak ada solusi yang memenuhi, maka c tidak merupakan kombinasi linier dari a dan b.

### 2. Diketahui u = (1,2), v = (2,2), w=(1,3). 
a. Apakah u, v dan w membangun R²?
b. Apakah u, v dan w bebas linier?

#### Langkah-langkah:
1. **Membuktikan Membangun R²**:
   - Untuk menentukan apakah u, v, dan w membangun R², kita harus mengecek apakah vektor-vektor tersebut linear independen.
   - Jika vektor-vektor tersebut linear independen, maka mereka membangun R².

2. **Membuktikan Linear Independen**:
   - Untuk menentukan apakah u, v, dan w linear independen, kita harus mengecek apakah sistem persamaan linear:
   $$
   x_1 u + x_2 v + x_3 w = 0
   $$
   memiliki solusi yang unik (yaitu solusi (0, 0, 0)).

3. **Membuat Sistem Persamaan Linear**:
   - Misalkan kita ingin mengecek apakah u, v, dan w linear independen.
   - Maka kita harus menyelesaikan sistem persamaan linear:
     $$
     x_1 (1, 2) + x_2 (2, 2) + x_3 (1, 3) = (0, 0)
     $$
     - Ini dapat diubah menjadi sistem persamaan:
       $$
       x_1 + 2x_2 + x_3 = 0
       $$
       $$
       2x_1 + 2x_2 + 3x_3 = 0
       $$

4. **Mengecek Solusi Unik**:
    - Jika sistem persamaan linear di atas memiliki solusi unik (yaitu solusi (0, 0, 0)), maka u, v, dan w linear independen.
    - Jika tidak ada solusi yang memenuhi, maka u, v, dan w tidak linear independen.

#### Contoh:

### 3. Diberikan SPL homogen:
p+2q+3r=0
p+2q-3r=0
p+ 2q+3r=0,

Tentukan basis ruang solusi (buktikan) dan tentukan dimensinya.

#### Langkah-langkah:
1. **Membuat Sistem Persamaan Linear**:
   - Sistem persamaan linear yang diberikan adalah:
     $$
     p + 2q + 3r = 0
     $$
     $$
     p + 2q - 3r = 0
     $$
     $$
     p + 2q + 3r = 0
     $$

2. **Mengurangi Persamaan**:
   - Kita dapat mengurangi persamaan-persamaan untuk menghilangkan redundansi.
   - Misalkan kita mengurangi persamaan pertama dengan persamaan kedua:
     $$
     (p + 2q + 3r) - (p + 2q - 3r) = 0 - 0
     $$
     - Mengurangi:
       $$
       6r = 0
       $$
       - Maka, r = 0.

3. **Mengganti ke dalam Persamaan**:
   - Mengganti r = 0 ke dalam salah satu persamaan, misalkan persamaan pertama:
     $$
     p + 2q + 3(0) = 0
     $$
     - Mengurangi:
       $$
       p + 2q = 0
       $$

4. **Membuat Basis Ruang Solusi**:
   - Basis ruang solusi dapat diambil sebagai vektor yang memenuhi sistem persamaan linear.
   - Misalkan kita ingin menentukan basis ruang solusi.
   - Maka, kita dapat menulis:
     $$
     p = -2q
     $$
     - Maka, basis ruang solusi dapat diambil sebagai vektor:
       $$
       (p, q) = (-2q, q)
       $$

5. **Mengubah Basis ke Bentuk Standar**:
   - Basis ruang solusi dapat diubah ke bentuk standar dengan menghilangkan redundansi.
   - Misalkan kita ingin menentukan basis ruang solusi dalam bentuk standar.
   - Maka, kita dapat menulis:
     $$
     (p, q) = (-2q, q)
     $$
     - Maka, basis ruang solusi dalam bentuk standar dapat diambil sebagai vektor:
       $$
       (p, q) = (-2, 1)
       $$

6. **Menghitung Dimensi Ruang Solusi**:
    - Dimensi ruang solusi dapat dihitung dengan menghitung jumlah basis ruang solusi.
    - Misalkan kita ingin menentukan dimensi ruang solusi.
    - Maka, kita dapat menulis:
      $$
      dimensi = 1
      $$
