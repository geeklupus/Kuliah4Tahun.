1. 
a. Determinan P dapat dihitung dengan mengalikan elemen-elemen diagonal utama matriks P dan menjumlahkannya. Dalam hal ini, determinan P adalah $2*6 - 1*4 = 8$.

b. Invers P dapat dihitung dengan menggunakan rumus:
$P^{-1} = \frac{1}{\text{det}(P)} \text{adj}(P)$
Di mana $\text{det}(P)$ adalah determinan P dan $\text{adj}(P)$ adalah adjoint dari P. Adjoint dari matriks P dapat dihitung dengan mengalikan elemen-elemen diagonal utama matriks P dan menjumlahkannya. Dalam hal ini, adjoint dari P adalah $[\begin{matrix} 6&-1\\ 4&2\end{matrix} ]$. Jadi, invers P adalah:
$P^{-1} = \frac{1}{8} [\begin{matrix} 6&-1\\ 4&2\end{matrix} ]$

a. Determinan A dapat dihitung dengan mengalikan elemen-elemen diagonal utama matriks A dan menjumlahkannya. Dalam hal ini, determinan A adalah $2*1 - 0*3 + 1*2 = 2$.

b. Adjoint A dapat dihitung dengan mengalikan elemen-elemen diagonal utama matriks A dan menjumlahkannya. Dalam hal ini, adjoint A adalah $[\begin{matrix} 1&-2&1\\ 2&-1&2\\ 3&2&1\end{matrix} ]$.

c. Invers A dapat dihitung dengan menggunakan rumus:
$A^{-1} = \frac{1}{\text{det}(A)} \text{adj}(A)$
Di mana $\text{det}(A)$ adalah determinan A dan $\text{adj}(A)$ adalah adjoint dari A. Jadi, invers A adalah:
$A^{-1} = \frac{1}{2} [\begin{matrix} 1&-2&1\\ 2&-1&2\\ 3&2&1\end{matrix} ]$
