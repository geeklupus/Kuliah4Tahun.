$[\begin{matrix} -1&0&3\\ 4&2&5\end{matrix} ][\begin{matrix} 3&1\\ 4&-2\\ 0&-3\end{matrix} ]$ adalah operasi perkalian matriks. Matriks pertama adalah $[\begin{matrix} -1&0&3\\ 4&2&5\end{matrix} ]$ dan matriks kedua adalah $[\begin{matrix} 3&1\\ 4&-2\\ 0&-3\end{matrix} ]$. Operasi perkalian matriks dilakukan dengan mengalikan elemen-elemen matriks tersebut dan menjumlahkan hasilnya. Dalam hal ini, hasil perkalian matriks tersebut adalah $[\begin{matrix} 6, 7, 0\end{matrix} ]$.
