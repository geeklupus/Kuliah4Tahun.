Untuk menyelesaikan tugas proyeksi orthogonal sesuai dengan pertanyaan yang diberikan, kita perlu menggunakan konsep proyeksi skalar ortogonal. Berikut adalah langkah-langkah untuk masing-masing pertanyaan:

### 01. Diketahui a=(1,k,4) dan b=(3,4,0), maka nilai k yang memenuhi agar proyeksi skalar ortogonal a pada b adalah 1/5.

#### Langkah-langkah:
1. **Hitung Proyeksi Skalar Ortogonal**:
   - Proyeksi skalar ortogonal $$ a $$ pada $$ b $$ dapat dihitung dengan menggunakan rumus:
   $$
   \text{proj}_b(a) = \frac{a \cdot b}{|b|^2} b
   $$
   - Di sini, $$ a \cdot b = (1, k, 4) \cdot (3, 4, 0) = (1)(3) + (k)(4) + (4)(0) = 3 + 4k $$
   - $$ |b|^2 = (3)^2 + (4)^2 + (0)^2 = 9 + 16 = 25 $$
   - Jadi, proyeksi skalar ortogonal $$ a $$ pada $$ b $$ adalah:
   $$
   \frac{3 + 4k}{25} (3, 4, 0)
   $$

2. **Atur untuk 1/5**:
   - Kita ingin $$ \frac{3 + 4k}{25} = \frac{1}{5} $$
   - Mengatur persamaan:
   $$
   \frac{3 + 4k}{25} = \frac{1}{5}
   $$
   - Mengalikan kedua sisi dengan 25 untuk menghilangkan denominasi:
   $$
   3 + 4k = 5
   $$
   - Mengurangi 3 dari kedua sisi:
   $$
   4k = 2
   $$
   - Membagi kedua sisi dengan 4:
   $$
   k = \frac{2}{4} = \frac{1}{2}
   $$

### 02. Diketahui A=(1,2,3), B=(-1,2,-3) dan C=(3,2,1) merupakan titik-titik pada ruang XYZ, maka panjang proyeksi orthogonal vector terhadap vector adalah...

#### Langkah-langkah:
1. **Tentukan Vektor AB dan AC**:
   - Vektor AB adalah:
   $$
   AB = B - A = (-1, 2, -3) - (1, 2, 3) = (-2, 0, -6)
   $$
   - Vektor AC adalah:
   $$
   AC = C - A = (3, 2, 1) - (1, 2, 3) = (2, 0, -2)
   $$

2. **Hitung Proyeksi Skalar Ortogonal**:
   - Proyeksi skalar ortogonal $$ AB $$ pada $$ AC $$ dapat dihitung dengan menggunakan rumus:
   $$
   \text{proj}_{AC}(AB) = \frac{AB \cdot AC}{|AC|^2} AC
   $$
   - Di sini, $$ AB \cdot AC = (-2, 0, -6) \cdot (2, 0, -2) = (-2)(2) + (0)(0) + (-6)(-2) = -4 + 12 = 8 $$
   - $$ |AC|^2 = (2)^2 + (0)^2 + (-2)^2 = 4 + 4 = 8 $$
   - Jadi, proyeksi skalar ortogonal $$ AB $$ pada $$ AC $$ adalah:
   $$
   \frac{8}{8} (2, 0, -2) = (2, 0, -2)
   $$

3. **Panjang Proyeksi**:
   - Panjang proyeksi orthogonal adalah:
   $$
   \sqrt{(2)^2 + (0)^2 + (-2)^2} = \sqrt{4 + 4} = \sqrt{8} = 2\sqrt{2}
   $$

### 03. Diketahui A(1,-1,2), B(2,1,-1) dan C(1,0,-3) merupakan titik-titik di R³, maka tentukan Proyeksi vector Orthogonal terhadap ruas garis.

#### Langkah-langkah:
1. **Tentukan Vektor AB dan AC**:
   - Vektor AB adalah:
   $$
   AB = B - A = (2, 1, -1) - (1, -1, 2) = (1, 2, -3)
   $$
   - Vektor AC adalah:
   $$
   AC = C - A = (1, 0, -3) - (1, -1, 2) = (0, 1, -5)
   $$

2. **Tentukan Vektor Garis**:
   - Vektor garis dapat dihitung sebagai vektor yang tegak lurus dengan vektor AB dan AC. Kita bisa menggunakan rumus:
   $$
   \text{Vektor Garis} = AB \times AC
   $$
   - Di sini, $$ AB \times AC = (1, 2, -3) \times (0, 1, -5) $$
   - Menghitung cross product:
   $$
   (1, 2, -3) \times (0, 1, -5) = (2(-5) - (-3)(1), (-3)(0) - (1)(-5), (1)(1) - (2)(0)) = (-10 + 3, 0 + 5, 1 - 0) = (-7, 5, 1)
   $$

3. **Proyeksi Orthogonal terhadap Vektor Garis**:
   - Untuk menentukan proyeksi orthogonal terhadap vektor garis, kita perlu menghitung proyeksi skalar ortogonal dari vektor AB ke vektor garis.
   - Rumusnya adalah:
   $$
   \text{proj}_{\text{Vektor Garis}}(AB) = \frac{AB \cdot \text{Vektor Garis}}{|\text{Vektor Garis}|^2} \text{Vektor Garis}
   $$
   - Di sini, $$ AB \cdot \text{Vektor Garis} = (1, 2, -3) \cdot (-7, 5, 1) = (1)(-7) + (2)(5) + (-3)(1) = -7 + 10 - 3 = 0 $$
   - $$ |\text{Vektor Garis}|^2 = (-7)^2 + (5)^2 + (1)^2 = 49 + 25 + 1 = 75 $$
   - Jadi, proyeksi skalar ortogonal $$ AB $$ pada vektor garis adalah:
   $$
   \frac{0}{75} (-7, 5, 1) = (0, 0, 0)
   $$

Dengan demikian, proyeksi orthogonal vector terhadap vektor garis adalah vektor nol, karena proyeksi skalar ortogonalnya adalah nol.
