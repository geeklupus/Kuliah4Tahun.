Jawaban

Untuk menentukan nilai konstanta $$ a $$ sehingga sistem persamaan linear memiliki solusi tunggal, kita harus memastikan bahwa sistem persamaan linear tersebut memiliki sistem matriks yang invertibel. Berikut adalah langkah-langkah untuk menentukan nilai $$ a $$:

1. **Tulis Sistem Persamaan dalam Bentuk Matriks Aksiomatik:**

   $$
   \begin{cases}
   x + 2y - 3z = 4 \\
   3x - y + 5z = 2 \\
   4x + y + (a^2 - 14)z = a + 2
   \end{cases}
   $$

2. **Bentuk Matriks Aksiomatik:**

   $$
   \mathbf{A} = \begin{pmatrix}
   1 & 2 & -3 \\
   3 & -1 & 5 \\
   4 & 1 & a^2 - 14
   \end{pmatrix}
   $$

3. **Kondisi untuk Solusi Tunggal:**

   Untuk sistem persamaan linear memiliki solusi tunggal, matriks $$\mathbf{A}$$ harus memiliki determinan yang tidak nol ($$\det(\mathbf{A}) \neq 0$$).

4. **Mencari Determinan Matriks A:**

   $$
   \det(\mathbf{A}) = 
   \begin{vmatrix}
   1 & 2 & -3 \\
   3 & -1 & 5 \\
   4 & 1 & a^2 - 14
   \end{vmatrix}
   $$

5. **Menghitung Determinan dengan Menggunakan Metode Saringan:**

   - **Baris Pertama:**
     - $$1 \times (-1) - 2 \times (5) + (-3) \times (1) = -1 -10 -3 = -14$$

   - **Baris Kedua:**
     - $$3 \times (1) - (-1) \times (-3) + (5) \times (4) = 3 -3 +20 =20$$

   - **Baris Ketiga:**
     - $$4 \times (2) - (1) \times (-3) + (a^2 -14) \times (1)=8+3+(a^2-14)=a^2-3$$

6. **Menggabungkan Hasilnya:**

   $$
   \det(\mathbf{A}) =
   \begin{vmatrix}
   1 & 2 & -3 \\
   3 & -1 & 5 \\
   4 & 1 & a^2 -14
   \end{vmatrix}
   =
   (-14)(20)-(20)(a^2-3)=(-280)-(20a^2+60)=(-340)-(20a^2)
   $$

7. **Kondisi untuk Determinan Tidak Nol:**

    Untuk sistem memiliki solusi tunggal, kita harus memastikan bahwa determinan tidak nol, yaitu:

    $$
    (-340)-(20a^2)\neq0
    $$

8. **Mengatasi Persamaan:**

    $$
    (20a^2)+340=0
    $$

    $$
    a^2=\frac{-340}{20}
    $$

    $$
    a^2=-17
    $$

    Karena $$a^2$$ tidak bisa negatif, maka tidak ada nilai $$a$$ yang memenuhi kondisi ini.

9. **Kesimpulan:**

    Karena tidak ada nilai $$a$$ yang memenuhi kondisi determinan tidak nol, maka sistem persamaan linear tidak memiliki solusi tunggal. Oleh karena itu, sistem persamaan linear memiliki **tidak ada solusi**.

### Jawaban Soal Diskusi
