Untuk menentukan invers matriks $$ B $$ dengan menggunakan transformasi baris elementer, kita harus mengikuti langkah-langkah berikut:

1. **Buat Matriks Augmentasi:**
   - Matriks $$ B $$ diberikan sebagai:
     $$
     B = \begin{pmatrix}
     1 & 1 \\
     3 & 4 \\
     3 & 4 \\
     2 & 3
     \end{pmatrix}
     $$
   - Matriks augmentasi $$ [B | I] $$ adalah:
     $$
     [B | I] = \begin{pmatrix}
     1 & 1 & | & 1 & 0 \\
     3 & 4 & | & 0 & 1 \\
     3 & 4 & | & 0 & 0 \\
     2 & 3 & | & 0 & 0
     \end{pmatrix}
     $$

2. **Transformasi Baris Elementer untuk Membuat Matriks Identitas:**
   - **Langkah 1:**
     - Ubah baris pertama menjadi baris dominan dengan membagi baris pertama dengan 1.
     - Matriks menjadi:
       $$
       [B | I] = \begin{pmatrix}
       1 & 1 & | & 1 & 0 \\
       3 & 4 & | & 0 & 1 \\
       3 & 4 & | & 0 & 0 \\
       2 & 3 & | & 0 & 0
       \end{pmatrix}
       $$
   - **Langkah 2:**
     - Eliminasi elemen di bawah 1 di baris pertama dengan melakukan operasi baris elementer.
       - $$ R_2' = R_2 - 3R_1 $$
       - $$ R_3' = R_3 - 3R_1 $$
       - $$ R_4' = R_4 - 2R_1 $$
     - Matriks menjadi:
       $$
       [B | I] = \begin{pmatrix}
       1 & 1 & | & 1 & 0 \\
       0 & 1 & | & -2 & 1 \\
       0 & 1 & | & -2 & 0 \\
       0 & 1 & | & -1 & 0
       \end{pmatrix}
       $$
   - **Langkah 3:**
     - Eliminasi elemen di bawah 1 di baris kedua dengan melakukan operasi baris elementer.
       - $$ R_3' = R_3 - R_2' $$
       - $$ R_4' = R_4 - R_2' $$
     - Matriks menjadi:
       $$
       [B | I] = \begin{pmatrix}
       1 & 1 & | & 1 & 0 \\
       0 & 1 & | & -2 & 1 \\
       0 & 0 & | & 0 & -1 \\
       0 & 0 & | & 1 & -1
       \end{pmatrix}
       $$
   - **Langkah 4:**
     - Eliminasi elemen di bawah 1 di baris ketiga dengan melakukan operasi baris elementer.
       - $$ R_4' = R_4 + R_3' $$
     - Matriks menjadi:
       $$
       [B | I] = \begin{pmatrix}
       1 & 1 & | & 1 & 0 \\
       0 & 1 & | & -2 & 1 \\
       0 & 0 & | & 0 & -1 \\
       0 & 0 & | & 1 & 0
       \end{pmatrix}
       $$

3. **Mengidentifikasi Invers Matriks:**
   - Setelah melakukan transformasi baris elementer, kita dapat mengidentifikasi invers matriks $$ B $$ dari matriks augmentasi.
   - Invers matriks $$ B^{-1} $$ adalah:
     $$
     B^{-1} = \begin{pmatrix}
     1 & -1 & 0 & 0 \\
     -2 & 1 & 0 & 0 \\
     0 & -1 & 0 & 1 \\
     1 & -1 & -1 & 0
     \end{pmatrix}
     $$
