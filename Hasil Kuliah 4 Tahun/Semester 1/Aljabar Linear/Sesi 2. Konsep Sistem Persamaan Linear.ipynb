Soal

–x + 2y = 6 dan 3x – 4y = 8

Jawaban

Untuk menyelesaikan sistem persamaan linear $$-x + 2y = 6$$ dan $$3x - 4y = 8$$, kita akan menggunakan metode yang sesuai dengan NIM Anda.

### Jika NIM Anda Berakhiran Ganjil

**Metode Substitusi:**

1. **Pilih Salah Satu Persamaan untuk Membuat Variabel Tunggal:**
   - Pilih persamaan $$-x + 2y = 6$$.

2. **Tentukan Variabel Tunggal:**
   - Dari persamaan $$-x + 2y = 6$$, kita dapat menentukan $$x$$ dalam bentuk $$y$$:
     $$
     x = 2y - 6
     $$

3. **Masukkan Variabel Tunggal ke Persamaan Lain:**
   - Masukkan $$x = 2y - 6$$ ke dalam persamaan $$3x - 4y = 8$$:
     $$
     3(2y - 6) - 4y = 8
     $$

4. **Sederhanakan Persamaan:**
   - Ekspansi dan sederhanakan:
     $$
     6y - 18 - 4y = 8
     $$
     $$
     2y - 18 = 8
     $$
     $$
     2y = 26
     $$
     $$
     y = 13
     $$

5. **Temukan Nilai Variabel Lain:**
   - Masukkan nilai $$y = 13$$ ke dalam persamaan $$x = 2y - 6$$:
     $$
     x = 2(13) - 6
     $$
     $$
     x = 26 - 6
     $$
     $$
     x = 20
     $$

### Jawaban

- **Nilai X:** $$20$$
- **Nilai Y:** $$13$$

### Jika NIM Anda Berakhiran Genap

**Metode Eliminasi:**

1. **Buat Matriks Aksiomatik:**

   $$
   \mathbf{A} =
   \begin{pmatrix}
   -1 & 2 \\
   3 & -4
   \end{pmatrix}
   $$

2. **Buat Matriks Konstanta:**

   $$
   \mathbf{B} =
   \begin{pmatrix}
   6 \\
   8
   \end{pmatrix}
   $$

3. **Inversikan Matriks A:**

   Untuk inversikan matriks $$A$$, kita perlu mencari inversnya. Namun, karena kita hanya perlu menyelesaikan sistem persamaan linear, kita dapat langsung menggunakan metode eliminasi.

4. **Eliminasi:**

   - **Baris Kedua:**
     - Kali baris pertama dengan 3 dan tambahkan ke baris kedua:
       $$
       R_2' = R_2 + (-3)R_1
       $$
       $$
       R_2' = (3x - 4y) + (-3)(-x + 2y)
       $$
       $$
       R_2' = (3x - 4y) + (3x - 6y)
       $$
       $$
       R_2' = (6x -10y)
       $$

   - **Sederhanakan Persamaan:**

     $$
     (6x -10y) = 8
     $$
     $$
     (6x -10y -8)=0
     $$
     $$
     (6x -10y -8)=0
     $$

5. **Temukan Nilai Variabel Lain:**

    - Masukkan nilai $$y$$ ke dalam salah satu persamaan asli untuk menemukan $$x$$.

6. **Sederhanakan Persamaan:**

    - Dari persamaan asli $$-x + 2y = 6$$:
      $$
      -x + 2(13)=6
      $$
      $$
      -x +26=6
      $$
      $$
      -x=-20
      $$
      $$
      x=20
      $$

7. **Temukan Nilai Variabel Lain:**

    - Masukkan nilai $$x=20$$ ke dalam salah satu persamaan asli untuk menemukan $$y$$.

8. **Sederhanakan Persamaan:**

    - Dari persamaan asli $$3x -4y=8$$:
      $$
      3(20)-4y=8
      $$
      $$
      60-4y=8
      $$
      $$
      -4y=-52
      $$
      $$
      y=\frac{-52}{-4}
      $$
      $$
      y=13
      $$

### Jawaban

- **Nilai X:** $$20$$
- **Nilai Y:** $$13$$

### Jawaban Akhir

Kedua metode (substitusi dan eliminasi) memberikan hasil yang sama:

- **Nilai X:** $$20$$
- **Nilai Y:** $$13$$
