a. Untuk menentukan nilai k, kita dapat membandingkan sistem persamaan linear yang diberikan dengan bentuk matriks yang diperoleh dari sistem persamaan linear tersebut. Sistem persamaan linear yang diberikan dapat ditulis dalam bentuk matriks sebagai:
$[\begin{matrix} 1&2&-3\\ 3&-1&5\\ 4&1&(k^{2}-14)\end{matrix} ]x = [\begin{matrix} 0\\ 0\\ 0\end{matrix} ]$
Matriks ini dapat ditulis dalam bentuk matriks yang diperoleh dari sistem persamaan linear sebagai:
$[\begin{matrix} 1&2&-3\\ 3&-1&5\\ 4&1&(k^{2}-14)\end{matrix} ]x = [\begin{matrix} 0\\ 0\\ 0\end{matrix} ]$
Karena sistem persamaan linear ini memiliki jumlah variabel yang sama dengan jumlah persamaan, maka sistem ini memiliki solusi tak hingga banyak jika dan hanya jika matriks ini adalah matriks yang diperoleh dari sistem persamaan linear tersebut. Dengan demikian, kita dapat menentukan nilai k dengan membandingkan sistem persamaan linear yang diberikan dengan bentuk matriks yang diperoleh dari sistem persamaan linear tersebut.

b. Untuk menentukan solusi dari sistem persamaan linear yang diberikan, kita dapat menggunakan metode substitusi atau metode eliminasi. Metode substitusi melibatkan memecahkan salah satu variabel dalam istilah variabel lain dan kemudian menggantikan variabel tersebut dalam persamaan lainnya. Metode eliminasi melibatkan mengalikan salah satu persamaan dengan koefisien sehingga koefisien dari variabel yang sama dalam kedua persamaan menjadi sama, dan kemudian menambahkan kedua persamaan tersebut untuk menghilangkan variabel tersebut. Metode apa pun yang digunakan, hasilnya akan menjadi solusi dari sistem persamaan linear yang diberikan.
