Untuk menentukan vektor yang tegak lurhadap vektor $\bar{u} = (-7, 3, 1)$ dan $\bar{v} = (2, 0, 4)$, kita dapat menggunakan rumus:

$\bar{w} = \bar{u} - (\bar{u} \cdot \bar{v}) \frac{\bar{v}}{\vert \bar{v} \vert^2}$
Di mana $\bar{u} \cdot \bar{v}$ 

adalah hasil perkalian dot dari 

$\bar{u}$ dan $\bar{v}$, dan $\vert \bar{v} \vert^2$ adalah panjang kuadrat dari $\bar{v}$. 

Dengan mengganti nilai-nilai yang telah kita hitung, kita dapatkan:

$\bar{w} = (-7, 3, 1) - ((-7)(2) + (3)(0) + (1)(4)) \frac{(2, 0, 4)}{(2)^2 + (0)^2 + (4)^2}$
