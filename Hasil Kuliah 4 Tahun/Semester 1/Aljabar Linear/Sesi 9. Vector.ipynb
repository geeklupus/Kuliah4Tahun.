Untuk menggambar vektor posisi dari OA yang berpangkal di titik O, kita dapat menggunakan panah yang mengarah dari titik O ke titik A. Panah tersebut akan mewakili arah dan panjang vektor posisi dari OA.

3.

a. Untuk menyatakan vektor PQ dalam bentuk vektor kolom, kita dapat menulis vektor tersebut sebagai kolom dari komponen-komponennya. Misalnya, jika vektor PQ memiliki komponen x = 2 dan komponen y = 3, maka vektor PQ dapat disatakan sebagai:
$PQ = [\begin{matrix} 2\\ 3\end{matrix} ]$

b. Untuk menyatakan vektor PQ dalam bentuk i,j (vektor satuan), kita dapat menggunakan notasi vektor satuan. Misalnya, jika vektor PQ memiliki komponen x = 2 dan komponen y = 3, maka vektor PQ dapat disatakan sebagai:
$PQ = 2i + 3j$

c. Untuk menentukan modulus atau panjang vektor PQ, kita dapat menggunakan rumus:
$|PQ| = \sqrt{PQ_x^2 + PQ_y^2}$
Di mana PQ_x dan PQ_y adalah komponen-komponen vektor PQ. Dengan mengganti nilai-nilai yang telah kita hitung, kita dapatkan:
$|PQ| = \sqrt{2^2 + 3^2} = \sqrt{4 + 9} = \sqrt{13}$

4.

Untuk menentukan panjang vektor yang berpangkal di titik A, kita dapat menggunakan rumus yang sama seperti yang digunakan untuk menentukan modulus vektor PQ. Misalnya, jika vektor yang berpangkal di titik A memiliki komponen x = 2 dan komponen y = 3, maka panjang vektor tersebut dapat dihitung sebagai:
$|AA| = \sqrt{AA_x^2 + AA_y^2}$
Di mana AA_x dan AA_y adalah komponen-komponen vektor yang berpangkal di titik A. Dengan mengganti nilai-nilai yang telah kita hitung, kita dapatkan:
$|AA| = \sqrt{2^2 + 3^2} = \sqrt{4 + 9} = \sqrt{13}$
