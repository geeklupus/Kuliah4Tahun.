Jawaban

### 1. Judul: 7 Alasan Doa Belum Terkabulkan dalam Kristen

Judul yang sesuai untuk materi ini adalah: **"7 Alasan Doa Belum Terkabulkan dalam Kristen"**.

### 2. Materi yang Didapatkan

Materi yang didapatkan dari sumber-sumber yang diberikan adalah sebagai berikut:

1. **Dosa dan Niat Jahat**:
   - Doa tidak dikabulkan karena dosa atau adanya niat jahat dalam hati (Kristen Alkitabiah)[1].

2. **Maksiat dan Kurang Beramal**:
   - Doa tidak dikabulkan karena seseorang selalu berada dalam maksiat atau kurang bersedekah (Indah Shaliha)[2].

3. **Belum Waktunya dan Rencana Allah**:
   - Doa tidak dikabulkan karena waktu Tuhan belum tiba atau karena Tuhan memiliki rencana yang lebih baik (Kristen Alkitabiah, REC)[1][3].

4. **Lupa Bersyukur dan Membicarakan Kekurangan Orang Lain**:
   - Doa tidak dikabulkan karena seseorang lupa bersyukur atau selalu membicarakan kekurangan orang lain (Indah Shaliha)[2].

5. **Kurang Bersungguh-Sungguh dalam Doa**:
   - Doa tidak dikabulkan karena seseorang kurang bersungguh-sungguh dalam doa (Warungsatekamu)[4].

### 3. Berikan Contoh Bersosialisasi

Contoh bersosialisasi yang relevan dengan tema doa dan pengabulannya adalah:

**Contoh:**

- **Membantu Orang Lain dalam Doa**:
  - Seseorang Kristen yang sedang menghadapi kesulitan dapat meminta doa dari teman-temannya. Dengan demikian, mereka dapat bersosialisasi dan saling mendukung dalam iman mereka.
  - **Contoh Konkrit:**
    - Seorang teman Kristen yang sedang menghadapi kesulitan keuangan dapat meminta doa dari komunitas gereja. Dengan demikian, mereka dapat saling mendukung dan bersosialisasi dalam iman mereka.

### 4. Apa yang Paling Menarik dari Kotbah

Apa yang paling menarik dari kotbah tentang doa dan pengabulannya adalah:

**Menarik:**

- **Pesan tentang Iman yang Sederhana**:
  - Kotbah yang menekankan pentingnya iman yang sederhana dan jujur dalam doa. Misalnya, seperti yang dikatakan dalam Yakobus 5:16b, "Doa orang benar, bila dengan yakin didoakan, sangat besar kuasanya."
  - **Contoh Konkrit:**
    - Seorang pengkhotbah dapat memberikan contoh dari kisah-kisah Alkitab, seperti kisah Elia, yang menunjukkan bahwa fokus utama dari doa Kristen bukanlah untuk itu, melainkan untuk kemuliaan Allah (Warungsatekamu)[4].

Dengan demikian, kotbah yang menekankan iman yang sederhana dan jujur dalam doa akan menarik perhatian audiens dan memberikan mereka motivasi untuk terus berdoa dengan tekun.
